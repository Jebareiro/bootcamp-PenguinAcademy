```python
# Diccionarios
# Un diccionario es una colecci√≥n de pares formados por una clave
#  y un valor asociado a la clave. 
# Se construyen poniendo los pares entre llaves { } separados por comas, 
# y separando la clave del valor con dos puntos :.

# Se caracterizan por:

# No tener orden.
# Pueden contener elementos de distintos tipos.
# Son mutables, es decir, pueden alterarse durante la ejecuci√≥n de un programa.
# Las claves son √∫nicas, es decir, no pueden repetirse en un mismo diccionario, y pueden ser de cualquier tipo de datos.
```

```python
# Un diccionario es una estructura que guarda datos en pares de clave y valor.
# La clave (key) es como el "nombre".
# El valor (value) es el dato asociado a ese nombre.

mi_contacto = { 
    "nombre" : "Jessica",            # -> clave: "nombre"   valor: 'Jessica'
    "apellido" : "Bareiro",
    "celular" : "0987654321",
    "edad" : 22,
}
print(mi_contacto)
```

```python
#Desplegar datos del diccionario

print(mi_contacto.keys())  # Devuelve solo las claves del diccionario (nombre, apellido, etc.)
print('') 

print(mi_contacto.values())  # Devuelve solo los valores ('laura', 'Gomez', etc.)
print('')   

print(mi_contacto.items())   # Devuelve una lista de tuplas con todos los pares clave:valor
```

```python
# Acceder a llos datos del diccionario 
mi_contacto["nombre"]
mi_contacto["apellido"]

# Actualizar los datos del diccionario
mi_contacto["direccion"] = "Sacreamento - Edif DOT"
print(mi_contacto);

# Modificar Datos 
mi_contacto['celular'] = 9987654321
print(mi_contacto);

# Eliminar ultimo par
mi_contacto.popitem()
print(mi_contacto);

# Eliminar datos por su clave
del mi_contacto["edad"]
print(mi_contacto);
```

```python
# Challenge 1 ü§∫
# Escribir un programa que cree un diccionario vac√≠o y lo vaya llenado con informaci√≥n 
# sobre una persona (por ejemplo nombre, edad, sexo, tel√©fono, correo electr√≥nico, etc.)
#  que se le pida al usuario.
#  Cada vez que se a√±ada un nuevo dato debe imprimirse el contenido del diccionario.

diccionario = {}
nombre = input("Cual es tu nombre?")
diccionario["nombre"] = nombre
print(diccionario["nombre"]);

apellido = input("Cual es tu apellido?")
diccionario["apellido"] = apellido
print(diccionario["apellido"]);

edad = int(input("Cual es tu edad?"))
diccionario["edad"] = edad
print(diccionario["edad"]);

print(diccionario);
```

```python
# Cargar datos en un diccionario a partir de entradas del usuario,
#  repitiendo hasta que decida detenerse

mi_diccionario = {}

# crear informacion que se le pida al usuario
while True:
    clave = input("Que campo deseas agregar? (ej. nombre - edad): ")
    valor = input("Ingrese el valor para el diccionario")


    mi_diccionario[clave] = valor
    print(" Lista de Diccionario")
    print(mi_diccionario)

    continuar = input("Deseas agregar mas datos? si o no..").lower()
    if continuar != "si":
        break

    print(mi_diccionario)
```

```python
# Challenge 2 ü§∫
# Crear una tarjeta personal. El usuario (mediante input) puede elegir:

# Crear datos nuevos
# Modificar datos
# Eliminar datos

tarjeta = {} # tarjeta vacia

while True:
    eleccion = input("Que desea realizar? seleccione el numero correspondiente -  1) CREAR - 2) MODIFICAR -  3) ELIMINAR  4) VER DATOS - u otro digito para salir")

    # 1 CREAR DATOS NUEVOS
    if eleccion == "1":
        while True:
            campo = input("Ingrese el nombre del campo (ej. nombre, edad): ") #campo es la clave 
            valor = input("Ingrese el valor para ese campo: ")
            tarjeta[campo] = valor  # agrega el dato 
            print("Datos actuales:", tarjeta)

            continuar = input("¬øDesea agregar otro dato? (si o no): ").lower()
            if continuar != "si":
                break

    # 2 MODIFICAR DATOS EXISTENTES
    elif eleccion == "2":
        if tarjeta:
            tarjeta.keys()
            campo = input("Ingrese el campo que desea modificar: ")
            nuevo_valor = input("Ingrese el nuevo valor: ")
            tarjeta[campo] = nuevo_valor
            print("Datos actuales:", tarjeta)

    # 3 ELIMINAR DATOS
    elif eleccion == "3":
        if tarjeta:
            tarjeta.keys()
            campo = input("Ingrese el campo que desea eliminar: ")            
            del tarjeta[campo]
            print("Datos actuales:", tarjeta)

    # 4 VER TODOS LOS DATOS 
    elif eleccion == "4":
        if tarjeta:
            print("Datos guardados en la tarjeta:", tarjeta)
        else:
            print("La tarjeta no tiene ning√∫n dato.")

    # 5 Salir
    else:
        break 
```

```python
# Challenge 3 ü§∫
# Crear una agenda de contactos. El usuario puede elegir:

# Mostrar los contactos
# Crear contacto
# Modificar datos del contacto

agenda = {"contactos": {}}

while True:
    eleccion = input("\n¬øQu√© desea realizar? Ingrese el n√∫mero:\n1) CREAR\n2) MODIFICAR\n3) ELIMINAR\n4) VER DATOS\nOtro n√∫mero para salir: ")

    if eleccion == "1":  # CREAR DATO NUEVO
        while True:
            campo = input("Ingrese el nombre del campo del contacto (ej. nombre, edad): ")
            valor = input("Ingrese el valor para ese campo: ")
            agenda["contactos"][campo] = valor
            print("Dato agregado.")

            continuar = input("¬øDesea agregar otro dato? (si/no): ").lower()
            if continuar != "si":
                break

    elif eleccion == "2":  # MODIFICAR DATO
        if agenda["contactos"]:
            campo = input("Ingrese el campo que desea modificar: ")
            if campo in agenda["contactos"]:
                nuevo_valor = input("Ingrese el nuevo valor: ")
                agenda["contactos"][campo] = nuevo_valor
                print("Dato modificado.")
            else:
                print("Ese campo no existe.")
        else:
            print("No hay datos para modificar.")

    elif eleccion == "3":  # ELIMINAR CAMPO
        if agenda["contactos"]:
            campo = input("Ingrese el campo que desea eliminar: ")
            if campo in agenda["contactos"]:
                del agenda["contactos"][campo]
                print("Campo eliminado.")
            else:
                print("Ese campo no existe.")
        else:
            print("No hay datos para eliminar.")

    elif eleccion == "4":  # VER TODOS LOS DATOS
        if agenda["contactos"]:
            print("\nDatos guardados en la agenda:")
            for campo, valor in agenda["contactos"].items():
                print(f"{campo}: {valor}")
        else:
            print("No hay datos en la agenda.")

    else:
        print("Saliendo del programa.")
        break
```

```python
# Programaci√≥n orientada a objetos
# Python tambi√©n nos permite representar cosas del mundo real, m√°s fielmente
#  por medio de estructuras de datos llamadas objetos.

# Tanto sus atributos (caracter√≠sticas y propiedades) como sus m√©todos 
# (funciones que nos permiten interactuar con el objeto) se definen en una clase.

# Esta clase es como una plantilla para los objetos que queremos crear, como un molde.

# Luego ese molde se utiliza para crear la cantidad que necesitemos de los objetos de esa clase.
```

```python
class Perro :
    ojos = 2
    patas = 4

firu = Perro()
peter = Perro()

print(firu.ojos, firu.patas)
print(peter.ojos, peter.patas)
```

```python
# Definiendo atributos
# Es importante distinguir que existen dos tipos de atributos:

# Atributos de clase: Son atributos cuyos valores definen a la clase, 
# por lo tanto ser√°n comunes para todos los objetos. (Mostrar el que ya ten√≠amos arriba)

# Atributos de instancia: Son atributos cuyos valores son espec√≠ficos de los objetos creados (la instancia).
#  En nuestro caso, de ambas mascotas tiene el atributo nombre, pero cada una tiene un valor diferente,
#  que se define al crear el objeto. (Mostrar los de instancia)

class Perro:

#atributo de clase

    ojos = 2
    patas = 4

#atributo de instancia
    def __init__(self, color, raza):
        self.color = color
        self.raza = raza


# Crear varios perros distintos
yuwailer = Perro("marron", "delmer")
peter = Perro("blanco", "chihuahua")
lola = Perro("negro", "labrador")

# Imprimir sus datos
print(yuwailer.color, yuwailer.raza)  # marron delmer
print(peter.color, peter.raza)        # blanco chihuahua
print(lola.color, lola.raza)          # negro labrador



# ‚úÖ Atributo de clase: Es compartido por todos los objetos de esa clase.
# ‚úÖ Atributo de instancia: Es √∫nico para cada objeto.

# self es una referencia al propio objeto creado a partir de la clase.
# __init__() es el constructor de la clase
```

```python
# Challenge ü§∫
# Crear dos m√©todos para nuestro animal y ejecutar, debe ser algo espec√≠fico de ese animal.

# Ejemplo: "Cuando hablo, digo X". 
# Ejemplo: "Cuando me muevo, hago X".


class Perro:
    #atributo de clase = todo lo comun, todo lo que el objeto tiene de comun (SUSTANTIVO COMUN)
    ojos = 2


    #atributo de instancia, todo lo que le diferencia del resto, a un perro se le puede diferenciar por su raza o por su color (ADJETIVO CALIFICATIVO)
    def __init__(self, color, raza):
        self.color = color
        self.raza = raza #{init se usa solo una vez}


        #metodos, osea acciones que puede realizar el objeto (VERBO)
    def correr(self): 
        return('saltando')
    def ladrar(self):
        return('guau guau')

chiruzo = Perro('blanco', 'Caniche') #le creamos a chiruzo, acaba de nacer 


print(chiruzo.ladrar())
print(chiruzo.correr())
```

```python
# Herencia
# Ojo! Esto no es una herencia como tal, sino una relaci√≥n entre clases. 
# Lo que representa la herencia es que una clase hereda los atributos y m√©todos de otra clase. (mostrar ejemplo)

# Dado que una clase hija hereda los atributos y m√©todos de la padre,
#  nos puede ser muy √∫til cuando tengamos clases que se parecen entre s√≠ pero tienen ciertas particularidades

class Gato:
    orejas = 2
    patas = 4
    tipo = "Felino"

    def __init__(self, color, tamanho):
        self.color = color
        self.tamanho = tamanho

    def maulla(self):
        return "hola (en miau)"

    def sonido(self):
        return "Ronronea fuerte"

    def salta(self):
        return "salta alto"

# clase heredada definida fuera
class Gatito(Gato):
    cola = 1

    def jugar(self):
        return "Juega con la pelota"

# ahora s√≠, instanciamos bien
michi = Gatito("gris", "pequenho")

print(michi.maulla())
print(michi.jugar())
```

```python
# Challenge ü§∫
# Definir una clase madre que herede a sus clases hijas dos atributos de instancia. 
# Mostrar ambos atributos en pantalla desde los objetos de las clases hijas.

# Observaci√≥n: Editar las clases hijas para que no se inicialicen.

class Caballo():
    patas = 4
    ojos = 2
    orejas = 2
    boca = 1
    cola = 1
    nariz = 1

    def __init__(self, color, raza):

        self.color = color
        self.raza = raza

    def relinchar(self):
        return "Cuando hablo, digo HIIII"
    
    def galopar(self):
        return "Corro rapido por el campo"



class Pony(Caballo):
    tamanho = "Pequenho"
    fuerza = "Menos que un caballo grande"

    def jugar(self):
        return "Corre Saltando"


Tiro_al_blanco = Caballo("Marron", "Pura Sangre")
print(Tiro_al_blanco.relinchar())
print(Tiro_al_blanco.galopar())

print("")

Spirit = Pony("Blanco", "Shetland")
print(Spirit.jugar())
print(Spirit.relinchar())
print(Spirit.galopar())


```
